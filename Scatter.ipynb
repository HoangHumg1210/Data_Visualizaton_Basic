# Hướng dẫn toàn diện về Scatter Plot trong Matplotlib

## 1. Giới thiệu về Scatter Plot

**Scatter Plot** (biểu đồ phân tán) là một loại biểu đồ thống kê sử dụng các điểm để hiển thị giá trị của hai biến số. Vị trí của mỗi điểm trên trục ngang và trục dọc cho biết giá trị của từng biến.

### Tại sao Scatter Plot quan trọng?
- **Phát hiện mối quan hệ**: Giúp nhận biết mối tương quan giữa các biến
- **Phát hiện outliers**: Dễ dàng spotted các giá trị bất thường
- **Phân tích xu hướng**: Xác định pattern và trend trong dữ liệu
- **So sánh nhóm**: Phân biệt các nhóm dữ liệu khác nhau

## 2. Cú pháp cơ bản

```python
import matplotlib.pyplot as plt
import numpy as np

# Cú pháp đơn giản
plt.scatter(x, y)

# Cú pháp đầy đủ
plt.scatter(x, y, s=None, c=None, marker='o', cmap=None, 
           alpha=None, edgecolors=None, linewidths=None)
```

## 3. Các thuộc tính chính

### 3.1 Thuộc tính dữ liệu
| Thuộc tính | Mô tả | Ví dụ |
|------------|--------|-------|
| `x` | Tọa độ trục hoành (bắt buộc) | `[1, 2, 3, 4]` |
| `y` | Tọa độ trục tung (bắt buộc) | `[2, 4, 6, 8]` |

### 3.2 Thuộc tính kích thước và hình dạng
| Thuộc tính | Mô tả | Giá trị | Ví dụ |
|------------|--------|---------|-------|
| `s` | Kích thước điểm | Số hoặc array | `s=50` hoặc `s=[20, 40, 60]` |
| `marker` | Hình dạng điểm | String | `'o'`, `'s'`, `'^'`, `'d'` |

### 3.3 Thuộc tính màu sắc
| Thuộc tính | Mô tả | Giá trị | Ví dụ |
|------------|--------|---------|-------|
| `c` | Màu sắc | Color name, hex, array | `'red'`, `'#FF0000'`, `[1,2,3]` |
| `cmap` | Bản đồ màu | String | `'viridis'`, `'plasma'`, `'coolwarm'` |
| `vmin`, `vmax` | Giới hạn color mapping | Float | `vmin=0, vmax=100` |

### 3.4 Thuộc tính hiển thị
| Thuộc tính | Mô tả | Giá trị | Ví dụ |
|------------|--------|---------|-------|
| `alpha` | Độ trong suốt | 0.0 - 1.0 | `alpha=0.7` |
| `edgecolors` | Màu viền | Color | `'black'`, `'white'` |
| `linewidths` | Độ dày viền | Float | `linewidths=1.5` |

## 5. Color Maps (Bản đồ màu) phổ biến

### Sequential (Tuần tự)
- `viridis`: Tốt nhất cho dữ liệu tuần tự
- `plasma`: Màu sắc rực rỡ
- `inferno`: Từ đen đến vàng
- `magma`: Từ đen đến trắng

### Diverging (Phân kỳ)
- `coolwarm`: Xanh lạnh - đỏ ấm
- `RdYlBu`: Đỏ - vàng - xanh
- `seismic`: Tốt cho dữ liệu có điểm trung tâm

### Qualitative (Định tính)
- `tab10`: 10 màu phân biệt
- `Set1`: Màu sắc tương phản cao

## 6. Các tác dụng và ứng dụng của Scatter Plot

### 6.1 Phân tích mối tương quan
```python
# Tương quan dương
plt.scatter(height, weight)  # Chiều cao vs cân nặng

# Tương quan âm  
plt.scatter(study_time, error_rate)  # Thời gian học vs tỷ lệ lỗi

# Không tương quan
plt.scatter(shoe_size, iq_score)  # Size giày vs IQ
```

**Các loại tương quan:**
- **Tương quan dương mạnh**: Điểm tạo thành đường thẳng dốc lên
- **Tương quan âm mạnh**: Điểm tạo thành đường thẳng dốc xuống
- **Không tương quan**: Điểm phân tán ngẫu nhiên
- **Tương quan phi tuyến**: Điểm tạo thành đường cong

### 6.2 Phát hiện Outliers (Ngoại lệ)
Scatter plot giúp dễ dàng nhận diện các điểm dữ liệu bất thường:
- Điểm nằm xa khỏi xu hướng chung
- Giá trị cực đại hoặc cực tiểu bất thường
- Lỗi đo lường hoặc nhập liệu

### 6.3 Phân nhóm dữ liệu (Clustering)
```python
# Sử dụng màu sắc để phân biệt nhóm
plt.scatter(x[group1], y[group1], c='red', label='Nhóm 1')
plt.scatter(x[group2], y[group2], c='blue', label='Nhóm 2')
```

### 6.4 Phân tích đa chiều
Scatter plot có thể hiển thị đến 4 chiều dữ liệu:
- **X-axis**: Biến thứ nhất
- **Y-axis**: Biến thứ hai  
- **Size**: Biến thứ ba
- **Color**: Biến thứ tư

## 7. Ví dụ thực tế

## 8. Best Practices (Thực hành tốt)

### 8.1 Lựa chọn kích thước điểm
- **Ít dữ liệu (<50 điểm)**: `s=100-200`
- **Trung bình (50-500 điểm)**: `s=50-100`  
- **Nhiều dữ liệu (>500 điểm)**: `s=20-50`

### 8.2 Sử dụng transparency
```python
# Với dữ liệu ít
alpha=0.8

# Với dữ liệu nhiều hoặc chồng lấp
alpha=0.5-0.6
```

### 8.3 Kết hợp với grid
```python
plt.scatter(x, y)
plt.grid(True, alpha=0.3)  # Grid nhẹ giúp đọc giá trị
```

### 8.4 Thêm thông tin
```python
plt.scatter(x, y, s=size, c=colors, cmap='viridis')
plt.colorbar(label='Giá trị Z')  # Giải thích màu sắc
plt.xlabel('Biến X')
plt.ylabel('Biến Y') 
plt.title('Tiêu đề mô tả')
```

## 9. Các lỗi thường gặp và cách khắc phục

### 9.1 Overplotting (Điểm chồng lấp)
**Vấn đề**: Quá nhiều điểm chồng lên nhau
**Giải pháp**:
- Giảm `alpha` xuống 0.5-0.6
- Sử dụng `plt.hexbin()` thay vì scatter
- Jitter: thêm noise nhỏ vào dữ liệu

### 9.2 Kích thước không phù hợp
**Vấn đề**: Điểm quá to hoặc quá nhỏ
**Giải pháp**:
- Điều chỉnh tham số `s`
- Sử dụng `s=array*scale_factor` để scale

### 9.3 Màu sắc không phù hợp
**Vấn đề**: Màu khó phân biệt hoặc không thẩm mỹ
**Giải pháp**:
- Dùng colormap phù hợp với dữ liệu
- Kiểm tra accessibility cho người mù màu

## 11. Tóm tắt

Scatter Plot là công cụ mạnh mẽ cho:
- ✅ **Khám phá dữ liệu**: Hiểu mối quan hệ giữa các biến
- ✅ **Phát hiện pattern**: Xu hướng, clustering, outliers
- ✅ **Trực quan hóa đa chiều**: Tối đa 4 chiều trong một plot
- ✅ **So sánh nhóm**: Phân biệt categories khác nhau
- ✅ **Báo cáo**: Trình bày kết quả một cách trực quan

**Khi nào sử dụng Scatter Plot:**
- Có 2 biến liên tục cần phân tích mối quan hệ
- Muốn phát hiện outliers
- Cần so sánh nhiều nhóm dữ liệu
- Khám phá xu hướng trong dữ liệu

**Khi nào KHÔNG nên dùng:**
- Dữ liệu categorical (dùng bar plot)
- Time series (dùng line plot)
- Phân phối một biến (dùng histogram)
- Quá nhiều điểm (>10,000) mà không có pattern rõ ràng

````python
companies = ['A', 'B','C','D','E','F','G','H','I','J']
costs = [11,24, 16, 88, 100, 54, 79, 45, 81, 2 ]
revenues = [40, 50, 10, 30, 12, 60, 30, 70,11, 99]

df = pd.DataFrame({"Company": companies, "Costs": costs, 'Revenues': revenues})
df
````

````python
colors = []

for cost, revenune in zip(df['Costs'], df['Revenues']):
    if cost >= revenune:
        colors.append('red')
    else:
        colors.append('green')

# plt.axline((0,0), slope =1)
min_val = min(df['Costs'].min(), df['Revenues'].min())
max_val = max(df['Costs'].max(), df['Revenues'].max())
plt.plot((min_val, max_val), (min_val, max_val), linestyle = '--', color = 'gray')


plt.scatter(df['Costs'], df['Revenues'], color = colors, edgecolor = 'black')

for i, row in df.iterrows():
    plt.text(row['Costs'] +2, row['Revenues'], row['Company'])

plt.title("Mối quan hệ giữa chi phí và doanh thu của các công ty")
plt.xlabel("Chi phí (triệu đồng)")
plt.ylabel("Doanh thu (trieu dong)")
plt.grid()
````
