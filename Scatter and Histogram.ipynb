# Hướng dẫn toàn diện về Scatter Plot trong Matplotlib

## 1. Giới thiệu về Scatter Plot

**Scatter Plot** (biểu đồ phân tán) là một loại biểu đồ thống kê sử dụng các điểm để hiển thị giá trị của hai biến số. Vị trí của mỗi điểm trên trục ngang và trục dọc cho biết giá trị của từng biến.

### Tại sao Scatter Plot quan trọng?
- **Phát hiện mối quan hệ**: Giúp nhận biết mối tương quan giữa các biến
- **Phát hiện outliers**: Dễ dàng spotted các giá trị bất thường
- **Phân tích xu hướng**: Xác định pattern và trend trong dữ liệu
- **So sánh nhóm**: Phân biệt các nhóm dữ liệu khác nhau

## 2. Cú pháp cơ bản

```python
import matplotlib.pyplot as plt
import numpy as np

# Cú pháp đơn giản
plt.scatter(x, y)

# Cú pháp đầy đủ
plt.scatter(x, y, s=None, c=None, marker='o', cmap=None, 
           alpha=None, edgecolors=None, linewidths=None)
```

## 3. Các thuộc tính chính

### 3.1 Thuộc tính dữ liệu
| Thuộc tính | Mô tả | Ví dụ |
|------------|--------|-------|
| `x` | Tọa độ trục hoành (bắt buộc) | `[1, 2, 3, 4]` |
| `y` | Tọa độ trục tung (bắt buộc) | `[2, 4, 6, 8]` |

### 3.2 Thuộc tính kích thước và hình dạng
| Thuộc tính | Mô tả | Giá trị | Ví dụ |
|------------|--------|---------|-------|
| `s` | Kích thước điểm | Số hoặc array | `s=50` hoặc `s=[20, 40, 60]` |
| `marker` | Hình dạng điểm | String | `'o'`, `'s'`, `'^'`, `'d'` |

### 3.3 Thuộc tính màu sắc
| Thuộc tính | Mô tả | Giá trị | Ví dụ |
|------------|--------|---------|-------|
| `c` | Màu sắc | Color name, hex, array | `'red'`, `'#FF0000'`, `[1,2,3]` |
| `cmap` | Bản đồ màu | String | `'viridis'`, `'plasma'`, `'coolwarm'` |
| `vmin`, `vmax` | Giới hạn color mapping | Float | `vmin=0, vmax=100` |

### 3.4 Thuộc tính hiển thị
| Thuộc tính | Mô tả | Giá trị | Ví dụ |
|------------|--------|---------|-------|
| `alpha` | Độ trong suốt | 0.0 - 1.0 | `alpha=0.7` |
| `edgecolors` | Màu viền | Color | `'black'`, `'white'` |
| `linewidths` | Độ dày viền | Float | `linewidths=1.5` |

## 5. Color Maps (Bản đồ màu) phổ biến

### Sequential (Tuần tự)
- `viridis`: Tốt nhất cho dữ liệu tuần tự
- `plasma`: Màu sắc rực rỡ
- `inferno`: Từ đen đến vàng
- `magma`: Từ đen đến trắng

### Diverging (Phân kỳ)
- `coolwarm`: Xanh lạnh - đỏ ấm
- `RdYlBu`: Đỏ - vàng - xanh
- `seismic`: Tốt cho dữ liệu có điểm trung tâm

### Qualitative (Định tính)
- `tab10`: 10 màu phân biệt
- `Set1`: Màu sắc tương phản cao

## 6. Các tác dụng và ứng dụng của Scatter Plot

### 6.1 Phân tích mối tương quan
```python
# Tương quan dương
plt.scatter(height, weight)  # Chiều cao vs cân nặng

# Tương quan âm  
plt.scatter(study_time, error_rate)  # Thời gian học vs tỷ lệ lỗi

# Không tương quan
plt.scatter(shoe_size, iq_score)  # Size giày vs IQ
```

**Các loại tương quan:**
- **Tương quan dương mạnh**: Điểm tạo thành đường thẳng dốc lên
- **Tương quan âm mạnh**: Điểm tạo thành đường thẳng dốc xuống
- **Không tương quan**: Điểm phân tán ngẫu nhiên
- **Tương quan phi tuyến**: Điểm tạo thành đường cong

### 6.2 Phát hiện Outliers (Ngoại lệ)
Scatter plot giúp dễ dàng nhận diện các điểm dữ liệu bất thường:
- Điểm nằm xa khỏi xu hướng chung
- Giá trị cực đại hoặc cực tiểu bất thường
- Lỗi đo lường hoặc nhập liệu

### 6.3 Phân nhóm dữ liệu (Clustering)
```python
# Sử dụng màu sắc để phân biệt nhóm
plt.scatter(x[group1], y[group1], c='red', label='Nhóm 1')
plt.scatter(x[group2], y[group2], c='blue', label='Nhóm 2')
```

### 6.4 Phân tích đa chiều
Scatter plot có thể hiển thị đến 4 chiều dữ liệu:
- **X-axis**: Biến thứ nhất
- **Y-axis**: Biến thứ hai  
- **Size**: Biến thứ ba
- **Color**: Biến thứ tư

## 8. Best Practices (Thực hành tốt)

### 8.1 Lựa chọn kích thước điểm
- **Ít dữ liệu (<50 điểm)**: `s=100-200`
- **Trung bình (50-500 điểm)**: `s=50-100`  
- **Nhiều dữ liệu (>500 điểm)**: `s=20-50`

### 8.2 Sử dụng transparency
```python
# Với dữ liệu ít
alpha=0.8

# Với dữ liệu nhiều hoặc chồng lấp
alpha=0.5-0.6
```

### 8.3 Kết hợp với grid
```python
plt.scatter(x, y)
plt.grid(True, alpha=0.3)  # Grid nhẹ giúp đọc giá trị
```

### 8.4 Thêm thông tin
```python

plt.ylabel('Biến Y') 
plt.title('Tiêu đề mô tả')
```

## 9. Các lỗi thường gặp và cách khắc phục

### 9.1 Overplotting (Điểm chồng lấp)
**Vấn đề**: Quá nhiều điểm chồng lên nhau
**Giải pháp**:
- Giảm `alpha` xuống 0.5-0.6
- Sử dụng `plt.hexbin()` thay vì scatter
- Jitter: thêm noise nhỏ vào dữ liệu

### 9.2 Kích thước không phù hợp
**Vấn đề**: Điểm quá to hoặc quá nhỏ
**Giải pháp**:
- Điều chỉnh tham số `s`
- Sử dụng `s=array*scale_factor` để scale

### 9.3 Màu sắc không phù hợp
**Vấn đề**: Màu khó phân biệt hoặc không thẩm mỹ
**Giải pháp**:
- Dùng colormap phù hợp với dữ liệu
- Kiểm tra accessibility cho người mù màu

# Hướng dẫn toàn diện về Histogram trong Matplotlib

## 1. Giới thiệu về Histogram

**Histogram** (biểu đồ tần suất) là một biểu đồ thống kê dùng để hiển thị phân phối tần suất của một tập dữ liệu liên tục. Nó chia dữ liệu thành các khoảng (bins) và đếm số lượng quan sát trong mỗi khoảng.

### Histogram khác Bar Chart như thế nào?
| Histogram | Bar Chart |
|-----------|-----------|
| Dữ liệu liên tục | Dữ liệu rời rạc/categorical |
| Các cột liền kề nhau | Các cột có khoảng cách |
| Trục X là số | Trục X là category |
| Hiển thị phân phối | So sánh giá trị |

## 2. Tại sao Histogram quan trọng?

### 2.1 Các tác dụng chính
- **Hiểu phân phối dữ liệu**: Normal, skewed, bimodal, uniform
- **Phát hiện outliers**: Giá trị bất thường ở đuôi phân phối
- **Kiểm tra giả định**: Nhiều thuật toán ML yêu cầu dữ liệu normal
- **So sánh datasets**: Overlay nhiều histogram
- **Xác định ngưỡng**: Threshold cho classification
- **Quality control**: Kiểm tra sản phẩm có đạt tiêu chuẩn

### 2.2 Các loại phân phối phổ biến
```
Normal (Gaussian)     Skewed Right         Skewed Left
      /\                  /\                    /\
     /  \                /  \                  /  \
    /    \              /    \---             /    \
   /      \            /       \---       ---/      \
  /        \          /          \---  ---/          \

Bimodal              Uniform               Exponential
   /\  /\              ----                 /\
  /  \/  \            |    |               /  \---
 /        \           |    |              /     \---
/          \          |    |             /        \---
             \        |    |            /           \---
```

## 3. Cú pháp cơ bản

```python
import matplotlib.pyplot as plt
import numpy as np

# Cú pháp đơn giản
plt.hist(data)

# Cú pháp đầy đủ
plt.hist(data, bins=10, range=None, density=False, weights=None,
         cumulative=False, bottom=None, histtype='bar', align='mid',
         orientation='vertical', rwidth=None, log=False, color=None,
         label=None, alpha=None, edgecolor=None, linewidth=None)
```

## 4. Các thuộc tính chính

### 4.1 Thuộc tính dữ liệu cơ bản
| Thuộc tính | Mô tả | Giá trị | Ví dụ |
|------------|-------|---------|-------|
| `data` | Dữ liệu đầu vào (bắt buộc) | Array-like | `[1,2,3,4,5]` |
| `bins` | Số lượng hoặc vị trí bins | Int hoặc array | `bins=20`, `bins=[0,5,10,15]` |
| `range` | Phạm vi dữ liệu | Tuple | `range=(0, 100)` |

### 4.2 Thuộc tính hiển thị
| Thuộc tính | Mô tả | Giá trị | Ví dụ |
|------------|-------|---------|-------|
| `density` | Normalize thành density | Boolean | `density=True` |
| `cumulative` | Histogram tích lũy | Boolean | `cumulative=True` |
| `histtype` | Kiểu histogram | String | `'bar'`, `'step'`, `'stepfilled'` |
| `orientation` | Hướng hiển thị | String | `'vertical'`, `'horizontal'` |

### 4.3 Thuộc tính thẩm mỹ
| Thuộc tính | Mô tả | Giá trị | Ví dụ |
|------------|-------|---------|-------|
| `color` | Màu sắc | Color | `'blue'`, `'#FF0000'` |
| `alpha` | Độ trong suốt | 0.0-1.0 | `alpha=0.7` |
| `edgecolor` | Màu viền | Color | `'black'`, `'white'` |
| `linewidth` | Độ dày viền | Float | `linewidth=1.5` |
| `rwidth` | Độ rộng tương đối của bars | 0.0-1.0 | `rwidth=0.8` |

## 6. Lựa chọn số lượng Bins

### 6.1 Các quy tắc phổ biến
```python
import numpy as np

n = len(data)  # Số lượng data points

# Sturges' Rule
bins_sturges = int(np.ceil(np.log2(n) + 1))

# Square Root Rule  
bins_sqrt = int(np.ceil(np.sqrt(n)))

# Rice Rule
bins_rice = int(np.ceil(2 * n**(1/3)))

# Freedman-Diaconis Rule
q75, q25 = np.percentile(data, [75, 25])
iqr = q75 - q25
bin_width = 2 * iqr / (n**(1/3))
bins_fd = int(np.ceil((data.max() - data.min()) / bin_width))
```

### 6.2 Hướng dẫn thực tế
| Số lượng dữ liệu | Bins đề xuất | Lý do |
|------------------|--------------|-------|
| < 50 | 5-10 | Tránh noise |
| 50-200 | 10-20 | Cân bằng detail vs noise |
| 200-1000 | 20-50 | Hiển thị pattern tốt |
| > 1000 | 50-100 | Chi tiết cao |

## 9. Phân tích phân phối

```python
import scipy.stats as stats

# Tính các chỉ số
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)[0][0]
```

### 9.2 Kiểm tra tính normal
```python
# Overlay với normal distribution
plt.hist(data, bins=30, density=True, alpha=0.7, label='Data')

# Fit normal distribution
mu, sigma = stats.norm.fit(data)
```

## 10. Các loại phân phối đặc biệt

### 10.1 Phân phối Normal
**Đặc điểm:**
- Đối xứng quanh mean
- Mean = Median = Mode
- 68% dữ liệu trong ±1σ, 95% trong ±2σ

**Ví dụ:** Chiều cao, cân nặng, điểm IQ

### 10.2 Phân phối Skewed
**Right-skewed (Positive skew):**
- Đuôi dài bên phải
- Mean > Median > Mode
- Ví dụ: Thu nhập, giá nhà

**Left-skewed (Negative skew):**
- Đuôi dài bên trái  
- Mode > Median > Mean
- Ví dụ: Tuổi tại thời điểm nghỉ hưu

### 10.3 Phân phối Bimodal
**Đặc điểm:**
- Có 2 đỉnh
- Thường từ 2 nhóm khác nhau
- Ví dụ: Chiều cao nam + nữ

### 10.4 Phân phối Uniform
**Đặc điểm:**
- Tất cả giá trị có xác suất bằng nhau
- Hình chữ nhật
- Ví dụ: Số random, ngày sinh

### 11.3 Xử lý outliers
```python
# Loại bỏ outliers trước khi vẽ
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter data
filtered_data = data[(data >= lower_bound) & (data <= upper_bound)]
outliers = data[(data < lower_bound) | (data > upper_bound)]

plt.hist(filtered_data, bins=30, alpha=0.7, label='Normal data')
plt.hist(outliers, bins=10, alpha=0.7, color='red', label='Outliers')
plt.legend()
```

## 13. Kết hợp với Statistical Tests

### 13.1 Normality Tests
```python
from scipy import stats

# Shapiro-Wilk Test (n < 5000)
stat, p = stats.shapiro(data)
print(f'Shapiro-Wilk: p-value = {p:.4f}')

# Kolmogorov-Smirnov Test
stat, p = stats.kstest(data, 'norm', args=(np.mean(data), np.std(data)))
print(f'KS Test: p-value = {p:.4f}')

# Anderson-Darling Test
stat, critical_values, significance_level = stats.anderson(data, dist='norm')
print(f'Anderson-Darling: statistic = {stat:.4f}')
```

### 13.2 Goodness of Fit
```python
# Chi-square goodness of fit
observed_freq, bins = np.histogram(data, bins=20)
expected_freq = len(data) / len(observed_freq)  # Uniform expectation
chi2, p = stats.chisquare(observed_freq, expected_freq)
print(f'Chi-square test: p-value = {p:.4f}')
```

## 15. Tóm tắt

### 15.1 Khi nào sử dụng Histogram
✅ **Nên dùng khi:**
- Dữ liệu liên tục (continuous data)
- Muốn hiểu phân phối của một biến
- Kiểm tra tính normal của dữ liệu
- Phát hiện outliers
- So sánh phân phối của nhiều nhóm
- Exploratory Data Analysis (EDA)

❌ **Không nên dùng khi:**
- Dữ liệu categorical (dùng bar chart)
- Dữ liệu time series (dùng line plot)
- Muốn thấy mối quan hệ giữa 2 biến (dùng scatter plot)
- Dữ liệu quá ít (<20 điểm)

### 15.2 Checklist cho Histogram tốt
- [ ] Chọn số bins phù hợp
- [ ] Thêm xlabel, ylabel, title có ý nghĩa
- [ ] Xử lý outliers nếu cần
- [ ] Thêm thống kê quan trọng (mean, median)
- [ ] Sử dụng colors và alpha hợp lý
- [ ] Thêm grid nhẹ để dễ đọc
- [ ] Normalize nếu so sánh nhiều datasets

### 15.3 Các insight thường gặp

4. **Uniform**: Dữ liệu random hoặc được sample đều
