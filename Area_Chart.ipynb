Biểu đồ vùng: thể hiện sự thay đổi của 1 hoặc nhiều giá trị theo thời gian

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

months = np.arange(1,13)
sales_1 = np.array([200,220, 250, 270, 300, 350, 400, 420, 390, 350, 300, 250])
sales_2 = np.array([100,300, 450, 470, 320, 350, 400, 420, 390, 350, 300, 250])
```

```python
plt.figure(figsize=(8,6))
plt.stackplot(months, sales_1, sales_2,
              colors = ['orange', 'red'],
              labels = ['San_pham_A', "San_pham_B"],
             alpha = 0.5)

plt.title("So luong ban hang theo thang")
plt.yticks(range(0,1000, 200))
plt.xlabel("Thoi gian")
plt.ylabel('So luong')

plt.grid(axis = 'y')
plt.legend(loc = "upper left")
plt.show()
```

```python
# Tạo biểu đồ
plt.figure(figsize=(10, 4))
plt.title("Số lượng bán hàng theo tháng")
plt.xlabel('Tháng')
plt.ylabel("Số lượng bán hàng")

plt.fill_between(months, sales_1,
                 color = "blue",
                 where = (sales_1 > 200),
                alpha = 0.6, 
                linewidth = 3, 
                label = "San pham A")

plt.fill_between(months,
                sales_2,
                where = (sales_2 > 100),
                color = 'yellow',
                alpha = 0.6,
                linewidth = 3,
                label = 'San pham B')
# Thiết lập nhãn trục x
plt.xticks(months, ["Th1", 'Th2', 'Th3', "Th4", "Th5", 'Th6', 'Th7', 'Th8', 'Th9', 'Th10', 'Th11', 'Th12'])

# Hiển thị lưới và chú thích
plt.grid(axis = 'y')
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
months = np.arange(1, 13)
sales_1 = np.array([200, 220, 250, 270, 300, 350, 400, 420, 390, 350, 300, 250])
sales_2 = np.array([100, 300, 450, 470, 320, 350, 400, 420, 390, 350, 300, 250])

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.title("Biểu đồ fill_betweenx - So sánh sản phẩm theo tháng")
plt.xlabel("Số lượng bán hàng")
plt.ylabel("Tháng")

# Tô vùng giữa sales_1 và sales_2 theo chiều Y (months)
plt.fill_betweenx(months, sales_1, 
                  where=(300> sales_1),
                  color='lightgreen', alpha=0.5, label='Sản phẩm B > A')



# Vẽ đường biểu diễn
plt.plot(sales_1, months, label='Sản phẩm A', color='blue', linewidth=2)
plt.plot(sales_2, months, label='Sản phẩm B', color='orange', linewidth=2)

# Đổi nhãn trục Y thành tháng
plt.yticks(months, [f"Th{m}" for m in months])
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df = sns.load_dataset("flights")
df


```

```python
years = df['year'].unique()
months = np.arange(1,13)
alphas = np.linspace(0.2, 0.8, len(years))

plt.figure(figsize=(11,7))
for i, year in enumerate(years):
    data = df[df['year'] == year]['passengers']
    plt.fill_between(months, data, alpha = alphas[i], label = year, linewidth = 4)
    plt.plot(months, data, alpha = alphas[i], label = year)

plt.legend(loc = 1)
plt.show()
```
