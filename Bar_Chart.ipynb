```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
the_loai = ['Giáo dục', 'Giải trí', 'Thể thao', 'Tin tức', 'Du lịch']
luot_xem_youtube = [120, 300, 210, 90, 160]
luot_xem_tiktok =  [200, 350, 180, 110, 220]
```

```python
for i in range(len(the_loai)):
    print(f"{the_loai[i]}: YouTube = {luot_xem_youtube[i]}, TikTok = {luot_xem_tiktok[i]}")
```

```python
for i, view in enumerate(luot_xem_tiktok):
    print(f"{the_loai[i]}: {view} lượt xem TikTok")
```

```python
width = 0.3
x = np.arange(len(the_loai))

plt.bar(x - width / 2, luot_xem_tiktok, width = width, color = 'skyblue', edgecolor = 'black', label = 'Tiktok')
plt.bar(x + width /2, luot_xem_youtube, width= width, color = 'y', edgecolor = 'black', label = "Youtube")

for i in x:
    plt.axvline(x=i, color='red', linestyle='--')

for i in range(len(x)):
    plt.text(x[i] - width/2, luot_xem_tiktok[i] + 5, str(luot_xem_tiktok[i]), ha= 'center')
    plt.text(x[i] + width/2, luot_xem_youtube[i] + 5, str(luot_xem_youtube[i]), ha= 'center')
plt.legend()
plt.show()
```

```python
height = 0.3
y = np.arange(len(the_loai))

plt.figure(figsize=(10,6))

plt.barh(y, luot_xem_youtube, height = height, color = 'r', label = 'Youtube', edgecolor = "b")
for i in range(len(y)):
    plt.text(luot_xem_youtube[i] + 10, y[i], str(luot_xem_youtube[i]), ha = "center")
plt.ylabel("Thể loại")
plt.xlabel("Lượt xem ")
plt.legend()
```

```python

height = 0.25
y = np.arange(len(the_loai))

plt.figure(figsize=(10, 6))

# Vẽ 2 nhóm bar ngang
plt.barh(y - height/2, luot_xem_youtube, height=height, color='b', label='YouTube', edgecolor='black')
plt.barh(y + height/2, luot_xem_tiktok,  height=height, color='r', label='TikTok', edgecolor='black')

# Gắn giá trị trên thanh
for i in range(len(y)):
    plt.text(luot_xem_youtube[i] + 5, y[i] - height/2, str(luot_xem_youtube[i]), va='center', fontsize=9)
    plt.text(luot_xem_tiktok[i] + 5, y[i] + height/2,  str(luot_xem_tiktok[i]), va='center', fontsize=9)

# Cấu hình trục và tiêu đề
plt.yticks(y, the_loai)
plt.xlabel("Lượt xem (nghìn)")
plt.ylabel("Thể loại")
plt.title("So sánh lượt xem YouTube và TikTok theo thể loại")
plt.legend()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
width = 0.25
x = np.arange(len(the_loai))

plt.bar(x - width/2,luot_xem_youtube,width=width, color = 'b', label = "Youtube", edgecolor = 'black')
plt.bar(x + width/2, luot_xem_tiktok, width = width, color = "r", label = "Tik tok", edgecolor = 'black' )
for i in range(len(x)):
    plt.text(x[i] - width/2, luot_xem_youtube[i] +1,str(luot_xem_youtube[i]), ha='center')
    plt.text(x[i] + width/2, luot_xem_tiktok[i] + 1, str(luot_xem_tiktok[i]), ha="center")

plt.xticks(x, the_loai)
plt.xlabel("Thể loại")
plt.ylabel("Lượt xem (nghìn)")
plt.title("So sánh lượt xem YouTube và TikTok theo thể loại")
plt.legend()
plt.grid(axis='y', alpha=0.1)
plt.tight_layout()
plt.show()
```

### Bài tâp


```python
data = {"Month" : ['Janurary', 'February', 'March', 'April', "May"],
       'Sales': [120, 231, 432, 354, 588]}
df = pd.DataFrame(data)
df
```

```python
mean = df['Sales'].mean()
mean
```

```python
color = []
for v in (df['Sales']):
    if v > 400:
        color.append('skyblue')
    else:
        color.append('green')
plt.barh( df['Month'],df['Sales'], edgecolor = 'black', linewidth = 1.4, color = color)

# for i in range(len(df)):
#     plt.axhline(y=i, linestyle='--', color='gray', linewidth=0.5)
plt.axvline(x=mean, color='red', linestyle='--', linewidth=2, label=f'Trung bình: {round(mean, 1)}')

for i, value in enumerate(df['Sales']):
    plt.text(value + 5, i,str(i), va = 'center'  )

plt.title("Biểu đồ cột ngang giữa tháng và doanh số")
plt.xlabel("Doanh số")
plt.ylabel("Tháng")
plt.xlim(0, 800)
plt.legend()
plt.show()
```
