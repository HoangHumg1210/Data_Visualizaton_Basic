| Tham s·ªë         | Ki·ªÉu d·ªØ li·ªáu      | Gi·∫£i th√≠ch ng·∫Øn g·ªçn |
|------------------|-------------------|----------------------|
| `x`, `y`         | str               | T√™n c·ªôt d·ªØ li·ªáu tr·ª•c ho√†nh (X) v√† tung (Y) |
| `data`           | DataFrame         | D·ªØ li·ªáu ƒë·∫ßu v√†o ch·ª©a c√°c c·ªôt `x`, `y`, `hue`,... |
| `hue`            | str (optional)    | Nh√≥m ph√¢n lo·∫°i ƒë·ªÉ v·∫Ω ƒë∆∞·ªùng h·ªìi quy & ƒëi·ªÉm theo m√†u kh√°c nhau |
| `col`, `row`     | str (optional)    | T·∫°o l∆∞·ªõi bi·ªÉu ƒë·ªì con (FacetGrid) theo c·ªôt ho·∫∑c h√†ng |

| `palette`        | str ho·∫∑c dict     | B·∫£ng m√†u √°p d·ª•ng cho c√°c nh√≥m (`hue`) |
| `height`         | float             | Chi·ªÅu cao m·ªói bi·ªÉu ƒë·ªì con (inch) |

````python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('car_crashes')
df.head(3)
````

````python
sns.lmplot( x="alcohol",
           y="ins_premium",
           data=df,
           palette="Dark2",
           height=5,
           aspect=1
           )
plt.title("M·ªëi quan h·ªá gi·ªØa N·ªìng ƒë·ªô c·ªìn v√† Ph√≠ b·∫£o hi·ªÉm")
plt.xlabel("T·ª∑ l·ªá t√†i x·∫ø ·∫£nh h∆∞·ªüng b·ªüi c·ªìn (%)")
plt.ylabel("Ph√≠ b·∫£o hi·ªÉm($)")
````
