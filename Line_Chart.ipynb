### Biểu đồ Đường (Line Chart)
1. Cú pháp cơ bản
+ Biểu đồ đường đơn giản:
+ pythonimport matplotlib.pyplot as plt
+ import numpy as np

# Dữ liệu mẫu
+ x = np.linspace(0, 10, 100)
+ y = np.sin(x)

+ plt.plot(x, y)
+ plt.show()
+ Cú pháp chi tiết:
+ pythonplt.plot(x, y, fmt, **kwargs)
+ Tham số:

+ x: Mảng dữ liệu trục x
+ y: Mảng dữ liệu trục y
+ fmt: Chuỗi định dạng (màu, điểm đánh dấu, kiểu đường)
+ label: Nhãn cho chú giải
+ linewidth / lw: Độ dày đường
+ linestyle / ls: Kiểu đường ('-', '--', '-.', ':')
+ color / c: Màu sắc
+ marker: Kiểu điểm đánh dấu ('o', 's', '^', 'v', v.v.)
+ markersize / ms: Kích thước điểm đánh dấu
+ alpha: Độ trong suốt

+ 2. Nhiều biểu đồ đường
+ pythonimport matplotlib.pyplot as plt
+ import numpy as np

+ x = np.linspace(0, 10, 100)
+ y1 = np.sin(x)
+ y2 = np.cos(x)
+ y3 = np.sin(x + np.pi/4)

+ plt.figure(figsize=(10, 6))
+ plt.plot(x, y1, 'r-', label='sin(x)', linewidth=2)
+ plt.plot(x, y2, 'g--', label='cos(x)', linewidth=2)
+ plt.plot(x, y3, 'b:', label='sin(x+π/4)', linewidth=2)

+ plt.xlabel('Trục X')
+ plt.ylabel('Trục Y')
+ plt.title('Nhiều biểu đồ đường')
+ plt.legend()
+ plt.grid(True, alpha=0.3)
+ plt.show()
+ 3. Chuỗi định dạng
+ Màu sắc:

+ 'r': đỏ, 'g': xanh lá, 'b': xanh dương
- 'c': xanh lơ, 'm': tím đỏ, 'y': vàng, 'k': đen

- Kiểu đường:

- '-': đường liền
- '--': đường đứt nét
- '-.': đường chấm gạch
- ':': đường chấm

- Điểm đánh dấu:

+ 'o': hình tròn, 's': hình vuông, '^': tam giác lên
+ 'v': tam giác xuống, 'D': kim cương, '*': ngôi sao

+ Ví dụ chuỗi định dạng:
+ pythonplt.plot(x, y, 'ro-')    # Đỏ, điểm tròn, đường liền
+ plt.plot(x, y, 'g^--')   # Xanh lá, tam giác, đường đứt nét
+ plt.plot(x, y, 'bs:')    # Xanh dương, hình vuông, đường chấm
+ 4. Tạo kiểu nâng cao
+ Tùy chỉnh kiểu:
+ pythonplt.plot(x, y,
  +        color='#FF5733',        # Màu hex
    +      linestyle='--',         # Đường đứt nét
      +    linewidth=3,            # Độ dày
    +  marker='o',             # Điểm tròn
       +  markersize=8,           # Kích thước điểm
      +   markerfacecolor='blue', # Màu bên trong điểm
    + markeredgecolor='red',  # Màu viền điểm
+     markeredgewidth=2,      # Độ dày viền điểm
 +        alpha=0.7)              # Độ trong suốt
+ Subplot với nhiều biểu đồ:
+ pythonfig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Biểu đồ 1
+ ax1.plot(x, y1, 'r-', linewidth=2)
+ ax1.set_title('Sóng Sin')
+ ax1.grid(True)

# Biểu đồ 2
+ ax2.plot(x, y2, 'b--', linewidth=2)
+ ax2.set_title('Sóng Cos')
+ ax2.grid(True)

+ plt.tight_layout()
+ plt.show()
+ 5. Ví dụ thực tế
+ Dữ liệu chuỗi thời gian:
+ pythonimport matplotlib.pyplot as plt
+ import numpy as np
+ from datetime import datetime, timedelta

# Tạo dữ liệu thời gian
+ ngay_thang = [datetime.now() - timedelta(days=x) for x in range(30, 0, -1)]
+ nhiet_do = np.random.normal(25, 5, 30)  # Nhiệt độ trung bình 25°C

+ plt.figure(figsize=(12, 6))
+ plt.plot(ngay_thang, nhiet_do, 'o-', linewidth=2, markersize=6)
+ plt.title('Nhiệt độ theo thời gian', fontsize=16)
+ plt.xlabel('Ngày')
+ plt.ylabel('Nhiệt độ (°C)')
+ plt.xticks(rotation=45)
+ plt.grid(True, alpha=0.3)
+ plt.tight_layout()
+ plt.show()

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z = [1, 4, 9, 16, 25]


plt.plot(x, y,
         # marker = "o", 
         # linestyle = '--',
         'kx--',
         linewidth = 2,
         markersize = 8,
         mfc = 'white',
         mec = 'blue',
         alpha = 0.8,
         zorder = 2,
         label = "Du lieu 1"
       )

plt.title("Biểu đồ đường tùy chỉnh")
plt.xlabel("Thời gian")
plt.ylabel("Giá trị")
plt.grid(True)
plt.legend()
plt.show()
```

📝 ĐỀ BÀI: Biểu đồ so sánh doanh số bán hàng của 3 cửa hàng

🎯 Yêu cầu:
Vẽ biểu đồ đường cho 3 cửa hàng trên cùng một biểu đồ.

Mỗi đường phải có:

Màu khác nhau

Marker khác nhau

Kiểu đường khác nhau

Độ trong suốt khác nhau

Gắn nhãn (label) cho từng đường để dùng với legend().

Dùng markerfacecolor, markeredgecolor, markersize để tùy chỉnh marker.

Tăng độ dày đường (linewidth).

Gắn:

Tiêu đề biểu đồ

Nhãn trục X và Y

Lưới (grid)

Tăng kích thước biểu đồ với plt.figure(figsize=(...).

```python
thang = list(range(1, 13))  # 12 tháng

cuahang_A = [50, 60, 65, 80, 75, 90, 100, 110, 105, 95, 85, 80]
cuahang_B = [40, 55, 70, 60, 85, 95, 90, 100, 95, 85, 70, 60]
cuahang_C = [45, 50, 55, 70, 65, 85, 95, 105, 100, 90, 75, 65]
```

```python
plt.figure(figsize=(10,6))

plt.plot(thang, cuahang_A,
         'ro--',
         alpha = 0.5,
         label = "Cửa hàng A",
         markerfacecolor='white',
         mec = "b",
         zorder = 2,
         lw = 3
        )

plt.plot(thang, cuahang_B,
         'bx-.',
         label = "Cửa hàng B")

plt.plot(thang, cuahang_C,
         'ko-',
         alpha = 0.77,
         label = "Cửa hàng C")

plt.title("Biểu đồ đường")
plt.xlabel('Thời gian')
plt.ylabel("Cửa hàng")
plt.legend( fontsize = 9)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.plot(thang, cuahang_A, '--.', linewidth=2, label = "Cua hang A")
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 10))  # 3 hàng, 1 cột

# Cửa hàng A
axs[0].plot(thang, cuahang_A, 'ro--', label="Cửa hàng A", alpha=0.8)
axs[0].set_title("Cửa hàng A")
axs[0].legend()
axs[0].grid(True)

# Cửa hàng B
axs[1].plot(thang, cuahang_B, 'bx-.', label="Cửa hàng B", alpha=0.8)
axs[1].set_title("Cửa hàng B")
axs[1].legend()
axs[1].grid(True)

# Cửa hàng C
axs[2].plot(thang, cuahang_C, 'ko-', label="Cửa hàng C", alpha=0.8)
axs[2].set_title("Cửa hàng C")
axs[2].legend()
axs[2].grid(True)

# Gán nhãn trục chung (tùy chọn)
for ax in axs:
    ax.set_xlabel("Tháng")
    ax.set_ylabel("Doanh số")

plt.tight_layout()
plt.show()
```
