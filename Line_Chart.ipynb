```python
thang = list(range(1, 13))  # 12 th√°ng

cuahang_A = [50, 60, 65, 80, 75, 90, 100, 110, 105, 95, 85, 80]
cuahang_B = [40, 55, 70, 60, 85, 95, 90, 100, 95, 85, 70, 60]
cuahang_C = [45, 50, 55, 70, 65, 85, 95, 105, 100, 90, 75, 65]
```

```python
plt.figure(figsize=(10,6))

plt.plot(thang, cuahang_A,
         'ro--',
         alpha = 0.5,
         label = "C·ª≠a h√†ng A",
         markerfacecolor='white',
         mec = "b",
         zorder = 2,
         lw = 3
        )

plt.plot(thang, cuahang_B,
         'bx-.',
         label = "C·ª≠a h√†ng B")

plt.plot(thang, cuahang_C,
         'ko-',
         alpha = 0.77,
         label = "C·ª≠a h√†ng C")

plt.title("Bi·ªÉu ƒë·ªì ƒë∆∞·ªùng")
plt.xlabel('Th·ªùi gian')
plt.ylabel("C·ª≠a h√†ng")
plt.legend( fontsize = 9)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.plot(thang, cuahang_A, '--.', linewidth=2, label = "Cua hang A")
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 10))  # 3 h√†ng, 1 c·ªôt

# C·ª≠a h√†ng A
axs[0].plot(thang, cuahang_A, 'ro--', label="C·ª≠a h√†ng A", alpha=0.8)
axs[0].set_title("C·ª≠a h√†ng A")
axs[0].legend()
axs[0].grid(True)

# C·ª≠a h√†ng B
axs[1].plot(thang, cuahang_B, 'bx-.', label="C·ª≠a h√†ng B", alpha=0.8)
axs[1].set_title("C·ª≠a h√†ng B")
axs[1].legend()
axs[1].grid(True)

# C·ª≠a h√†ng C
axs[2].plot(thang, cuahang_C, 'ko-', label="C·ª≠a h√†ng C", alpha=0.8)
axs[2].set_title("C·ª≠a h√†ng C")
axs[2].legend()
axs[2].grid(True)

# G√°n nh√£n tr·ª•c chung (t√πy ch·ªçn)
for ax in axs:
    ax.set_xlabel("Th√°ng")
    ax.set_ylabel("Doanh s·ªë")

plt.tight_layout()
plt.show()
```
