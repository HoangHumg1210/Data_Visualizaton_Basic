## ‚ö° T√≥m t·∫Øt quan tr·ªçng

> üéØ **Ghi nh·ªõ:**
> - `plt.pie()` d√πng ƒë·ªÉ tr·ª±c quan t·ª∑ l·ªá ph·∫ßn trƒÉm
> - `autopct` gi√∫p hi·ªÉn th·ªã ph·∫ßn trƒÉm tr√™n bi·ªÉu ƒë·ªì
> - D√πng `explode` ƒë·ªÉ nh·∫•n m·∫°nh ph·∫ßn quan tr·ªçng
> - `plt.axis('equal')` ƒë·ªÉ gi·ªØ h√¨nh tr√≤n ƒë·ªÅu


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = {'Product': ['Chanh', "Dua hau", "Cam", "Chuoi", "Dao"],
       "Sales": [300, 123, 421, 100, 600]}

df = pd.DataFrame(data)
df
```

```python

plt.pie(df['Sales'], labels = df['Product'],
        colors= ['lightgreen', 'skyblue', 'red', 'lightcoral', 'yellow'],
        autopct= "%1.2f%%",startangle=180, explode=(0.2, 0,0,0,0), 
        shadow = True
        )
plt.legend(title = "San pham", loc = 'upper right', bbox_to_anchor= (1.4,1))
plt.title("Ti trong doanh so ban hang theo san pham")
plt.show()
```
